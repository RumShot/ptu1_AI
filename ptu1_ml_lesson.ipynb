```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
# mpg.info()
```

```python
mpg[mpg['horsepower'].isnull()]
```

```python
mpg['horsepower'].loc[32] = 75
mpg['horsepower'].loc[126] = 105
mpg['horsepower'].loc[330] = 51
mpg['horsepower'].loc[336] = 120
mpg['horsepower'].loc[354] = 73
mpg['horsepower'].loc[374] = 82
mpg
```

```python
mpg.info()
```

```python
origins = pd.get_dummies(mpg['origin'])
origins[100:115]
```

```python
data = pd.concat([mpg, origins], axis=1)
data.tail()
```

```python
data.drop(['origin','name'], axis=1, inplace=True)
```

```python
data[111:117]
```

```python
x = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'europe', 'japan', 'usa']]
# x = data[1:]
x
```

```python
y = data['mpg']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=24)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(x_train, y_train)
reg
```

```python
prediction = reg.predict(x_test)
prediction
```

```python
spejimai = pd.Series(data=prediction, name='Spejimas')
spejimai 
```

```python
rezultatas = pd.concat([y_test.reset_index(), spejimai], axis=1)
rezultatas
```

```python
reg.score(x_test,y_test)
```

```python
rezultatas['paklaida'] = rezultatas['mpg'] - rezultatas['Spejimas']
rezultatas
```

```python
sns.histplot(rezultatas['paklaida'])
```

```python
su_pavadinimais = pd.concat([mpg, rezultatas.set_index('index')], axis=1)
su_pavadinimais[su_pavadinimais['Spejimas'].notna()]
```
